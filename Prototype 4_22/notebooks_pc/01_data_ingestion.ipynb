### Extract Image Files 
[For direct links only]

```python
def extract_file_id(link):
    match = re.search(r'/d/([a-zA-Z0-9_-]+)', link)
    return match.group(1) if match else None

df = pd.read_excel("../data/test-image.xlsx")

for index, row in df.iterrows():
    file_id = extract_file_id(row["image_link"])
    
    if file_id:
        url = f"https://drive.google.com/uc?export=download&id={file_id}"
        
        response = requests.get(url, verify=False)
        
        if response.status_code == 200:
            with open(f"image_{index}.jpg", "wb") as f:
                f.write(response.content)
            print(f"Downloaded image_{index}.jpg")
        else:
            print("Failed:", index)
```

### Extract Image Files 
[For folder links]

```python
def extract_folder_id(link):
    match = re.search(r'/folders/([a-zA-Z0-9_-]+)', link)
    return match.group(1) if match else None

df = pd.read_excel("../data/test-image.xlsx")

for index, folder_link in enumerate(df["image_link"]):
    if pd.isna(folder_link):
        continue
    folder_id = extract_folder_id(folder_link)

    if folder_id:

        folder_url = f"https://drive.google.com/drive/folders/{folder_id}"
        output_folder = f"../data/images/folder_{index+1}"
        os.makedirs(output_folder, exist_ok=True)

        print(f"\nDownloading folder {index+1}...")

        try:
            gdown.download_folder(
                folder_url,
                output=output_folder,
                quiet=False
            )

            print("Success")
            time.sleep(5)

        except Exception as e:
            print("Error:", e)
            print("Waiting before retry...")
            time.sleep(15)

    else:
        print(f"No valid folder link in row {index}")
```

### Sort in Folders

```python
import os
import shutil
import pandas as pd

source_root = "../data/images"
target_root = "../data/img_data"

df = pd.read_excel("../data/test-image.xlsx")

os.makedirs(target_root, exist_ok=True)

counter = 0

for index, row in df.iterrows():

    folder_name = f"folder_{index+1}"
    folder_path = os.path.join(source_root, folder_name)

    status = row["status"]

    status_folder = os.path.join(target_root, status)
    os.makedirs(status_folder, exist_ok=True)

    if os.path.isdir(folder_path):

        for file in os.listdir(folder_path):

            if file.lower().endswith((".jpg", ".jpeg", ".png")):

                src = os.path.join(folder_path, file)
                dst = os.path.join(status_folder, f"img_{counter}.jpg")

                shutil.copy(src, dst)
                counter += 1

print("Done. Total images:", counter)
```
